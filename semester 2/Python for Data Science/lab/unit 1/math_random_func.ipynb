```python
import math
import random
# Math functions
print(math.sqrt(144))      # 12.0
print(math.pow(2, 10))     # 1024.0
print(math.floor(3.9))     # 3
print(math.ceil(3.1))      # 4
print(math.pi)             # 3.14159...
print(math.factorial(5))   # 120
print(math.log(100, 10))   # 2.0
```

```python
import math
import random
# Random functions
print(random.random())          # Float between 0 and 1
print(random.randint(1, 100))   # Random int between 1-100
print(random.choice([10, 20, 30, 40]))  # Random element
lst = [1, 2, 3, 4, 5]
random.shuffle(lst)
print(lst)                      # Shuffled list
print(random.sample(lst, 3))    # 3 unique random elements
```
